```python
import os
import random
import shutil
from PIL import Image
from glob import glob
```

```python
train_set = [i for i in glob('train/*')]
test_set = [i for i in glob('test/*')]
train_set
```

```python
normal_list = []
virus_list = []
for i in train_set:
    if 'NORMAL' in i:
        normal_list.append(i)
```

```python
normal_list
```

```python
for i in train_set:
    if i not in normal_list:
        virus_list.append(i)
```

```python
virus_list
```

```python
bact_list = []
for i in virus_list:
    if 'person' in i:
        bact_list.append(i)
```

```python
bact_list
```

```python
os.mkdir('Training/normal')
os.mkdir('Training/covid')
os.mkdir('Testing/normal')
os.mkdir('Testing/covid')
```

```python
print(f' len of normal set is :{len(normal_list)}\n len of effected list is {len(bact_list)}')
```

### NORMAL IMAGES

```python
original_list = []
train_list = []
for i in normal_list:
    img = Image.open(i)
    original_list.append(img)
```

```python
original_list
```

```python
for i in original_list:
    image = i.resize((224,224))
    train_list.append(image)
```

```python
for (i,new) in enumerate(train_list):
    new.save('{}{}{}'.format('Training/normal/',i+1,'.jpg'))
```

### COVID IMAGES

```python
original_list = []
covid_list = []
for i in bact_list:
    img = Image.open(i)
    original_list.append(img)
    
```

```python
for i in original_list:
    image = i.resize((224,224))
    covid_list.append(image)
```

```python
for (i,new) in enumerate(covid_list):
    new.save('{}{}{}'.format('Training/covid/',i+1,'.jpg'))
```

```python
source_path = 'Training/normal/'
dest_path = 'Testing/normal/'
img = os.listdir(source_path)
```

```python
random.shuffle(img)
```

```python
for j in range(0,100):
    images = img[j]
    img_path = os.path.join(source_path,images)
    dest_path1 = os.path.join(dest_path,images)
    shutil.copy2(img_path,dest_path1)
```
